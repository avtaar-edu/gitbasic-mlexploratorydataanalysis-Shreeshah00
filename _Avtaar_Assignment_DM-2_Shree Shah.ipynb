```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading all 5 dataset
car_1=pd.read_csv('assng2_dataset - Sheet1.csv')
car_2=pd.read_csv('assng2_dataset - Sheet2.csv')
car_3=pd.read_csv('assng2_dataset - Sheet3.csv')
car_4=pd.read_csv('assng2_dataset - Sheet4.csv')
car_5=pd.read_csv('US_Accidents_Dec20_Updated_sheet1.csv')
```

```python
# merge all datasets into one and in dtype as object
l = [car_1,car_2,car_3,car_4,car_5]
car_all = pd.concat(l)
```

```python
#Exclude unnecessary columns
exclude = ["End_Lat","End_Lng","Number","Timezone",
           "Airport_Code","Precipitation(in)","Weather_Timestamp"]
car_all = car_all.drop(exclude,axis=1)
car_all.head()
```

```python
car_all.isnull().sum()
```

## a. Which city had the most amount of accidents

## b. Identify if there was any time of the year or day that had more reported accidents? Try to infer why.


```python
car_all.head()
```

```python
# taking only time dataframe
data = pd.DataFrame(car_all[["Start_Time","Temperature(F)",'Humidity(%)','Pressure(in)','Visibility(mi)']])
data.head()
```

```python
data.isnull().sum()
```

```python
# spliting the date and time 
x = data["Start_Time"].str.split(" ", expand = True)

# adding date and time column in the dataframe
data['Date']=x[0]
data['Time']=x[1]

# removing the start_Time column from dataframe
data.drop(columns=["Start_Time"],inplace=True)

data
```

```python
# extracting month from date
data["Month"] = data["Date"].str[3:5]

# extracting hour from time and removing ':'
data["Hour"] = data["Time"].str[:2].replace({':':""},regex=True)


data
```

```python
# checking null values
data.isnull().sum()
```

```python
# droping rows with null values
data.dropna(inplace=True)
```

```python
# counting months to find number of accident in each month
count_month = data['Month'].value_counts()
count_month = pd.DataFrame(count_month)
count_month
```

```python
# ploting graph of month vs number of accident in that month
plt.plot(count_month)
plt.xlabel('MONTH')
plt.ylabel('Number of Accident')
plt.show
```

We can see that in December there is maximum car accident.

```python
# counting hour to find number of accident in that hour
count_hour = data['Hour'].value_counts()
count_hour = pd.DataFrame(count_hour)
count_hour
```

```python
# ploting graph of hour vs number of accident in that hour
plt.plot(count_hour)
plt.xlabel('Hour')
plt.ylabel('Number of Accident')
plt.show
```

At 8am and 7 am maximum number of accident occur 

```python
# changing the data type to integer and float
data['Temperature(F)']=pd.to_numeric(data['Temperature(F)'])
data['Humidity(%)']=pd.to_numeric(data['Humidity(%)'])
data['Pressure(in)']=pd.to_numeric(data['Pressure(in)'])
data['Visibility(mi)']=pd.to_numeric(data['Visibility(mi)'])
data['Month']=pd.to_numeric(data['Month'])
data['Hour']=pd.to_numeric(data['Hour'])
```

```python
data.head()
```

```python
# taking mean with respect to month
avg_month=data.groupby('Month').mean()
avg_month
```

```python
# ploting graph of month vs temperature
plt.plot(avg_month['Temperature(F)'])
plt.xlabel('MONTH')
plt.ylabel("Temperature(F)")
plt.show()
```

```python
# ploting graph of month vs Humidity
plt.plot(avg_month['Humidity(%)'])
plt.xlabel('MONTH')
plt.ylabel("Humidity(%)")
plt.show()
```

```python
# ploting graph of month vs Pressure
plt.plot(avg_month['Pressure(in)'])
plt.xlabel('MONTH')
plt.ylabel("Pressure(in)")
plt.show()
```

```python
# ploting graph of month vs Visibility
plt.plot(avg_month['Visibility(mi)'])
plt.xlabel('MONTH')
plt.ylabel("Visibility(mi)")
plt.show()
```

```python
import seaborn as sns
sns.heatmap(avg_month, annot=True, linewidths= .5,fmt= '.1f');
```

From the observation from above graphs, we can state that due low temperature and visibility and high humidity,the number of accident increase in december and november.

```python
# taking mean with respect to hour
avg_hour=data.groupby('Hour').mean()
avg_hour
```

```python
# ploting graph of month vs temperature
plt.plot(avg_hour['Temperature(F)'])
plt.xlabel('Hour')

plt.ylabel("Temperature(F)")
plt.show()
```

```python
# ploting graph of month vs Visibility
plt.plot(avg_hour['Visibility(mi)'])
plt.xlabel('Hour')
plt.ylabel("Visibility(mi)")
plt.show()
```

From the observation from above graphs, we can state that due low temperature, humidity and visibility and high humidity,the number of accident is more at 8am and 7am.
